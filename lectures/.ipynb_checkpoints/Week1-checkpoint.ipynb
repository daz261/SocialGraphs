```python
import networkx as nx
G =nx.Graph()
G
```

```python
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from([
    (4,{"color": "green"}),
    (5,{"color": "red"})
])

H = nx.path_graph(10)
G.add_nodes_from(H)
G.add_node(H)
```

```python
# Add an edge between node 1 and 2
G.add_edge(1,2)
# between 2 and 3
e = (2,3)
G.add_edge(*e) #unpack e, an edge tuple, with *
# Add the same 2 edges, but in one line
G.add_edges_from([(1,2),(1,3)])

# add edges from H to G
G.add_edges_from(H.edges)
# clear all nodes and edges
G.clear()
G.add_edges_from([(1,2),(1,3)])
G.add_node(1)
G.add_edge(1,2)
G.add_node("spam") # add node calles "spam"
G.add_nodes_from("spam") # adds 4 nodes:"s", "p", "a", "m"
G.add_edge(3,"m")
print(f'Number of nodes: {G.number_of_nodes()}')
print(f'Number of edges: {G.number_of_edges()}')

# order of edge addition
G.adj
```

```python
DG = nx.DiGraph()
DG.add_edge(2,1)
DG.add_edge(1,3)
DG.add_edge(2,4)
DG.add_edge(1,2)
assert list(DG.successors(2)) == [1,4]
assert list(DG.edges) == [(2,1),(2,4),(1,3),(1,2)]
DG.adj
print(f'DG nodes: {list(DG.nodes)}')
print(f'G nodes: {list(G.nodes)}')
print(f'G Edges: {list(G.edges)}')
print(f'G adjacencies index 1: {list(G.adj[1])}')
```

```python
print(G.edges([2,'m']))
print(G.degree([2,3]))
print(f'current nodes: {list(G.nodes)}')
G.remove_node(2)
G.remove_nodes_from("spam")
print(f'Remaining nodes: {list(G.nodes)}')
G.remove_edge(1,3)
```

```python
G.add_edge(1,2)
H = nx.DiGraph(G) # create new graph with nodes and edges from G
print(f'H edges: {H.edges}')
edgelist = [(0,1),(1,2),(2,3)]
H = nx.Graph(edgelist)
print(f'H edges: {H.edges}')
```

```python
# when in doubt, consider using convert_node_labels_to_integers() to obtain a more traditional graph with integer labels

G = nx.Graph([(1,2,{'color':'yellow'})])
print(G[1]) # same as G.adj[1]
print(G[1][2])
print(G.edges[1,2])

# set attribute of edges using subscripts
G.add_edge(1,3)
G[1][3]['color'] = 'blue'
print(f'Edge 1,3: {G.edges[1,3]}')
G.edges[1,2]['color'] = 'red'
print(f'Edge 1,2: {G.edges[1,2]}')
```

```python
FG = nx.Graph()

FG.add_weighted_edges_from([(1,2,0.125), (1,3,0.75), (2,4,1.2), (3,4,0.375)])
# print the added edges in a nice format
for n,nbrs in FG.adj.items():
    for nbr,eattr in nbrs.items():
        wt = eattr['weight']
        if wt < 0.5:
            print(f'({n},{nbr},{wt:.3})')
            
print('------------------------')
for (u,v,wt) in FG.edges.data('weight'):
    if wt < 0.5:
        print(f'({u},{v},{wt})')
```

```python
G = nx.Graph(day="Friday")
print(G.graph)

G.graph['day'] = 'Monday'
print(G.graph)
```

```python
G.add_node(1, time='5pm')
G.add_nodes_from([3], time='2pm')
print(G.nodes[1])
G.nodes[1]['room'] = 714
print(G.nodes.data())


G.add_edge(1, 2, weight=4.7 )
G.add_edges_from([(3, 4), (4, 5)], color='red')
G.add_edges_from([(1, 2, {'color': 'blue'}), (2, 3, {'weight': 8})])
G[1][2]['weight'] = 4.7
G.edges[3, 4]['weight'] = 4.2
G.edges[3,4]
```

```python
#Example of drawing a graph with matplotlib and draw / draw_shell command
import matplotlib.pyplot as plt

G = nx.petersen_graph()
subax1 = plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
subax2 = plt.subplot(122)
nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')
plt.show()

G = nx.dodecahedral_graph()

options = {
    'node_color': 'black',
    'node_size': 100,
    'width': 3
}
shells = [[2, 3, 4, 5, 6], [8, 1, 0, 19, 18, 17, 16, 15, 14, 7], [9, 10, 11, 12, 13]]
nx.draw_shell(G, nlist=shells, **options)
```
